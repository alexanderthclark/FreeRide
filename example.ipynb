# Taxes

```python
d = Demand.from_formula("p = 12-1*q")
s = Supply.from_formula("p = q")

m = Market(d,s)
m.tax = 2
ax = m.plot(surplus=True)
ax.set_title(f"DWL from Per-Unit Tax is ${m.dwl:.2f}")
plt.show()

print(m.producer_surplus)
```

# Subsidy

```python
d = Demand.from_formula("p = 12-1*q")
s = Supply.from_formula("p = q")

m = Market(d,s)
m.subsidy = 4
ax = m.plot(surplus=True)
ax.set_xlabel("Quantity of Commodity Corn")

ax.set_title(f"DWL from Per-Unit Subsidy is ${m.dwl:.2f}\nGovernment Expenditure is ${m.govt_expenditure:.2f}")
plt.show()

print(f"Consumer Surplus: ${m.consumer_surplus:.2f}")
print(f"Producer Surplus: ${m.producer_surplus:.2f}")
```

# Price Controls

```python
d = Demand.from_formula('P = 12 - 1*Q')
s = Supply.from_formula("P = Q")

m = Market(d, s) # perfectly competitive market

m.ceiling = 4
ax = m.plot(surplus=True)
ax.set_xlabel("Quantity of Housing")

plt.title(f"Deadweight loss from price ceiling = ${m.dwl:.2f}")
plt.show()

print(f"Consumer Surplus: ${m.consumer_surplus:.2f}")
print(f"Producer Surplus: ${m.producer_surplus:.2f}")
```

```python
d = Demand.from_formula('P = 12 - 1*Q')
s = Supply.from_formula("P = Q")

m = Market(d, s) # perfectly competitive market

m.floor = 10
ax = m.plot(surplus=True)
ax.set_xlabel("Quantity of Coffee Beans")

plt.title(f"Deadweight loss from price floor = ${m.dwl:.2f}")
plt.show()

print(f"Consumer Surplus: ${m.consumer_surplus:.2f}")
print(f"Producer Surplus: ${m.producer_surplus:.2f}")
```

# World Trade

```python
d = Demand.from_formula('P = 12 - 1*Q')
s = Supply.from_formula("P = Q")

# Small economy
m = Market(d, s)

# Open to world trade
m.world_price = 4

ax = m.plot(surplus=True)
ax.set_title("Open Economy")

print(f"Consumer Surplus: ${m.consumer_surplus:.2f}")
print(f"Producer Surplus: ${m.producer_surplus:.2f}")
print(f"Imports: {m.net_imports}" )
```

```python
d = Demand.from_formula('P = 12 - 1*Q')
s = Supply.from_formula("P = Q")

# Small economy
m = Market(d, s)

# Open to world trade
m.world_price = 4
m.tariff = 1

ax = m.plot(surplus=True)
ax.set_title(f"Equilibrium with Tariff\nDWL = ${m.dwl:.2f}")

print(f"Consumer Surplus: ${m.consumer_surplus:.2f}")
print(f"Producer Surplus: ${m.producer_surplus:.2f}")
print(f"Imports: {m.imports}" )
```

```python
d = Demand.from_formula('P = 12 - 1*Q')
s = Supply.from_formula("P = Q")

# Small economy
m = Market(d, s)

# Open to world trade
m.world_price = 4

m.tariff = 19421340921834
ax = m.plot(surplus=True)
ax.set_title(f"Prohibitive Tariff Eliminates All Imports\nDWL = ${m.dwl:.2f}")

print(f"Consumer Surplus: ${m.consumer_surplus:.2f}")
print(f"Producer Surplus: ${m.producer_surplus:.2f}")
print(f"Imports: {m.imports}" )
```
