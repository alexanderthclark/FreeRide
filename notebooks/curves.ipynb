# Affine

(piecewise affine)

```python
a = Affine(10, -2)
a
```

```python
a.plot()
```

```python
# horizontal sum
piecewise = Affine(intercept = (15,10), slope = (-1, -2))
piecewise
```

```python
# plotting isn't all there yet
piecewise.plot()
plt.gca().set_ylim(0,16)
```

```python
piecewise.pieces
```

```python
for piece in piecewise.pieces:
    if piece:
        print('q =', piece.expression)
        print('p =', piece.inverse_expression)
        print()
```
