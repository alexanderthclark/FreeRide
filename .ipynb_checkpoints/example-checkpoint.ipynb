```python
from IPython.display import clear_output
!pip install -e .
from microecon.curves import Demand, Supply
from microecon.equilibrium import Equilibrium
import matplotlib.pyplot as plt
clear_output()
```

# Create Demand Curves

```python
d1 = Demand.from_formula("Q = 10 - 1*P")
d2 = Demand.from_formula("P = 15 - 2*Q")
```

```python
d1
```

```python
d2
```

```python
d = d1 + d2

d
```

```python
d.plot()
```

# Create Supply Curves

```python
s1 = Supply.from_formula('Q = 2*P')
s2 = Supply.from_formula('Q = 2*P - 1')
```

```python
s = s1 + s2

s.plot()
```

# Equilibrium

```python
e = Equilibrium(d, s)

e.plot()
```

```python
# Labeled plot 

fig, ax = plt.subplots()
e.plot(ax = ax)
ax.set_ylim(0, 16)

ax.set_xlabel("Quantity")
ax.set_ylabel("Price")

# label curves
ax.text(2, d(2) + .05,
        s = 'D',
        size = 16,
        ha = 'left',
        va = 'bottom')

ax.text(16, s(16) + .05,
        s = 'S',
        size = 16,
        ha = 'right',
        va = 'bottom')

# label equilibrium
ax.plot([e.q], [e.p],
        marker = 'o',
        markersize = 10)

ax.text(e.q, e.p+2,
       s=f'Q={e.q:.2f}, P={e.p:.2f}',
       ha = 'center',
       va = 'bottom')

plt.savefig("equilibrium.svg", transparent = True)
plt.show()
```
